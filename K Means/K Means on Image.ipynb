```python
import numpy as np
```

```python
from PIL import Image
```

```python
import matplotlib.pyplot as plt
```

```python
image = Image.open('C:\\Users\\deept\\OneDrive\\Deep\\Bigdata\\deep.jpg')
```

```python
print(image.format)
print(image.size)
print(image.mode)
```

```python
img_data = np.asarray(image)
```

```python
np.shape(img_data)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(img_data)
plt.show()
```

```python
#Reshaping 3d matrix into 2 d as sklearn cannot take 3d matrix
data1 = img_data.reshape((-1,3))
```

```python
from  sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5, random_state=0).fit(data1)
```

```python
clusters = kmeans.labels_
```

```python
np.shape(clusters)
```

```python
colors = {0: (0.0, 0.75, 0.75), 1: (0.0, 0.0, 1.0), 2: (1.0, 1.0, 1.0), 3: (0.0, 0.5, 0.0), 4: (0.75, 0.75, 0), 5: (0.0, 0.0, 0.0), 6: (1.0, 0.0, 0.0), 7: (0.75, 0, 0.75)}
```

```python
a = np.zeros([87216,3])
j=0
for i in iter(kmeans.labels_):
    a[j] = np.array(colors[i])
    j +=1
```

```python
clusterimage = a.reshape(368, 237,3)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(clusterimage)
plt.show()
```
